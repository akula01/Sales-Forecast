```python
import pandas as pd
data = pd.read_csv('C:\Users\JBB\Documents\ESI 4628\Video Games Sales.csv')
data.head()
```

```python
df = data.iloc[:,10].mul(1000000)
df1 = pd.DataFrame(df, dtype=int)
del data['Global_Sales']
project_data = pd.merge(data, df1, left_index=True, right_index=True)
project_data.head()
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()

project_data['Platform'] = le.fit_transform(project_data['Platform'])
project_data['Genre'] = le.fit_transform(project_data['Genre'])
project_data['Publisher'] = le.fit_transform(project_data['Publisher'])
project_data['Year'] = le.fit_transform(project_data['Year'])
project_data.head()
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
```

```python
x = project_data[['Platform','Year','Genre','Publisher']].values
y = project_data[['Global_Sales']].values

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = 15)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

scaler.fit(x_train)

x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)
```

```python
clf = MLPClassifier(hidden_layer_sizes = (50,50,50), max_iter=500, alpha=0.0001, solver='sgd', random_state = 21, tol=0.0000001)
```

```python
clf.fit(x_train, y_train)
y_pred = clf.predict(x_test)
```

```python
score = accuracy_score(y_test, y_pred)
print(y_test)
print(y_pred)
print(score)
```
