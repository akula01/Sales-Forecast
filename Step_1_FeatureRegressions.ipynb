```python
import pandas as pd
import numpy as np
import datetime
from sklearn import svm
from matplotlib import pyplot as plt
from sklearn import preprocessing
```

```python
project_data = pd.read_csv('C:\Machine Learning Project\Video Games Sales.csv')
release_date = pd.read_csv('C:\Machine Learning Project\updatedrealeaseyear.csv')
release_date = release_date[release_date['Year'] != 'N/A  ']
```

```python
#convert global sales to integer

df1 = project_data.iloc[:,10].mul(1000000)
df2 = pd.DataFrame(df1, dtype=int)
del project_data['Global_Sales']
final_data = pd.merge(project_data, df2, left_index=True, right_index=True)
final_data.head()
```

```python
#encode data

le = preprocessing.LabelEncoder()

final_data['Platform'] = le.fit_transform(final_data['Platform'])
final_data['Genre'] = le.fit_transform(final_data['Genre'])
final_data['Publisher'] = le.fit_transform(final_data['Publisher'])

print(final_data.loc[29])
```

```python
#define test sets for each year (from 2014-2017)

df14 = (final_data['Year']>=2014) & (final_data['Year']<2015)
test_set_1 = final_data.loc[df14]

df15 = (final_data['Year']>=2015) & (final_data['Year']<2016)
test_set_2 = final_data.loc[df15]

df16 = (final_data['Year']>=2016) & (final_data['Year']<2017)
test_set_3 = final_data.loc[df16]

df17 = (final_data['Year']>=2017)
test_set_4 = final_data.loc[df17]
```

```python
#define training set

df_test = (final_data['Year'] <= 2013) & (final_data['Year'] >= 2005)
training_set = final_data.loc[df_test]
training_set.head()
```
