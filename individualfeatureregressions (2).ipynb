```python
import pandas as pd
import numpy as np
import datetime
from sklearn import metrics
from sklearn import svm
from sklearn.model_selection import train_test_split
from matplotlib import pyplot as plt
```

```python
csv_path = 'file:///Users/Stefan/Downloads/video_game_sales.csv'
project_data = pd.read_csv(csv_path)
csv_path1 = 'file:///Users/Stefan/Desktop/updatedrealeaseyear.csv'
release_date = pd.read_csv(csv_path1)
release_date = release_date[release_date['Year'] != 'N/A  ']
```

```python
df1 = project_data.iloc[:,10].mul(1000000)
df2 = pd.DataFrame(df1, dtype=int)
del project_data['Global_Sales']
final_data = pd.merge(project_data, df2, left_index=True, right_index=True)
final_data.head()
# converting the global sales value into the millions and updating the dtype to millions
```

```python
final_data.update(release_date)
final_data['Year'] = pd.to_datetime(final_data['Year'])
final_data.head()
```

```python
date1 = datetime.datetime(2013, 12, 31)
date2 = datetime.datetime(2005, 1, 1)
df = (final_data['Year'] <= date1) & (final_data['Year'] >= date2)
training_set = final_data.loc[df]
training_set
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()

training_set['Platform'] = le.fit_transform(training_set['Platform'])
training_set['Genre'] = le.fit_transform(training_set['Genre'])
training_set['Publisher'] = le.fit_transform(training_set['Publisher'])
training_set.head()
```

```python
date3 = datetime.datetime(2014, 12, 31)
date4 = datetime.datetime(2014, 1, 1)
df = (training_set['Year'] <= date3) | (training_set['Year'] >= date4)
test_set2014 = training_set.loc[df]

date5 = datetime.datetime(2015, 12, 31)
date6 = datetime.datetime(2015, 1, 1)
df1 = (training_set['Year'] <= date5) | (training_set['Year'] >= date6)
test_set2015 = training_set.loc[df1]

date7 = datetime.datetime(2016, 12, 31)
date8 = datetime.datetime(2016, 1, 1)
df2 = (training_set['Year'] <= date7) | (training_set['Year'] >= date8)
test_set2016 = training_set.loc[df2]

date9 = datetime.datetime(2017, 12, 31)
date10 = datetime.datetime(2017, 1, 1)
df3 = (training_set['Year'] <= date9) | (training_set['Year'] >= date10)
test_set2017 = training_set.loc[df3]
```

```python
#Defining Independent and Dependent Variables
fixed_platform_training_set = training_set.loc[training_set['Platform'] == 506]
fixed_platform_test_set = test_set2014.loc[test_set2014['Platform'] == 506]
X_train = fixed_platform_training_set.values[:,[3,4,5]]
X_test = fixed_platform_test_set.values[:,[3,4,5]]
y_train = fixed_platform_training_set.values[:,[10]]
y_test = fixed_platform_test_set.values[:,[10]]
```

```python
clf = svm.SVC(kernel='linear')
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```
